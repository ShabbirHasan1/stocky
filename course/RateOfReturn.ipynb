```python
import numpy  as np
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
```

```python
GSFC = wb.DataReader('GSFC.NS',data_source='yahoo',start = '2015-1-1')
```

```python
GSFC.head()
```

```python
GSFC["Simple Return"] = (GSFC["Adj Close"]/GSFC["Adj Close"].shift(1))-1
```

```python
print(GSFC["Simple Return"])
```

```python
GSFC["Simple Return"].plot(figsize=(8,5))
```

```python
avg_returns_d = GSFC["Simple Return"].mean()
avg_returns_d
```

```python
avg_returns_a = avg_returns_d*250
print(f"{round(avg_returns_a,5)*100}%")
```

```python
GSFC["Log Return"] = np.log(GSFC["Adj Close"]/GSFC["Adj Close"].shift(1))
print(GSFC["Log Return"])
```

```python
GSFC["Log Return"].plot(figsize=(8,5))
```

```python
avg_returns_d = GSFC["Log Return"].mean()
avg_returns_a = avg_returns_d*250
print(f"{round(avg_returns_a,5)*100}%")
```
